```python

import pandas as pd

file_path = r"D:\Depi\finalproject\planning\sample_-_superstore.xls"
df_source = pd.read_excel(file_path, sheet_name='source', engine='xlrd')
df_returns = pd.read_excel(file_path, sheet_name='Returns', engine='xlrd')
df_people = pd.read_excel(file_path, sheet_name='People', engine='xlrd')

df_final = pd.merge(df_source, df_returns, on='Order ID', how='left')

df_source.columns = df_source.columns.str.strip()
df_returns.columns = df_returns.columns.str.strip()

df_final = pd.merge(df_source, df_returns, on='Order ID', how='left')
df_final.head()

df_final = pd.merge(df_final, df_people, on='Region', how='left')

df_final['Returned'] = df_final['Returned'].fillna('No')


df_final['Returned Flag'] = (df_final['Returned'] == 'Yes').astype(int)



df_final['Order Date'] = pd.to_datetime(df_final['Order Date'])
df_final['Ship Date']  = pd.to_datetime(df_final['Ship Date'])

```
